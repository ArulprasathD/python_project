```python
import pandas as pd

data = pd.read_csv("hospital_patient_dataset.csv")
data = data.fillna(data.mean(numeric_only=True))

def risk(bp, sugar):
    if bp > 140 or sugar > 180:
        return "High"
    elif bp > 120 or sugar > 140:
        return "Medium"
    else:
        return "Low"

data["risk_level"] = data.apply(lambda row: risk(row["blood_pressure"], row["sugar_level"]), axis=1)

print(data[["name", "blood_pressure", "sugar_level", "risk_level"]].head())
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = pd.read_csv("hospital_patient_dataset.csv")
data = data.fillna(data.mean(numeric_only=True))

X = data[["age", "weight", "doctor_visits"]]
y = data["cholesterol"]

model = LinearRegression()
model.fit(X, y)

new_data = pd.DataFrame([[45, 72, 5]], columns=["age", "weight", "doctor_visits"])
prediction = model.predict(new_data)

print("Predicted cholesterol:", prediction[0])
```

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression

data = pd.read_csv("hospital_patient_dataset.csv")
data = data.fillna(data.mean(numeric_only=True))
data = data.fillna(data.mode().iloc[0])

data["recovered_num"] = data["recovered"].apply(lambda x: 1 if x == "Yes" else 0)

X = data[["age", "blood_pressure", "sugar_level", "doctor_visits"]]
y = data["recovered_num"]

model = LogisticRegression()
model.fit(X, y)

new_data = pd.DataFrame([[40, 130, 145, 4]], columns=["age", "blood_pressure", "sugar_level", "doctor_visits"])
result = model.predict(new_data)

if result[0] == 1:
    print("Recovered")
else:
    print("Not Recovered")
```
